## Programa√ß√£o orientada a objetos

* **Classes** s√£o estruturas que mesclam c√≥digo e dados e s√£o moldes para a constru√ß√£o de objetos.

* **Objetos** s√£o inst√¢ncias das classes que podem ser executadas e armazenam dados. Os objetos s√£o criados a partir das classes, recebem mensagens para que realizem alguma a√ß√£o.

Uma analogia que permite entender os conceitos de **classe** e **objetos** √© o que acontece quando voc√™ segue uma receita de bolo.

A receita de bolo define o que √© o bolo, mas a receita, apesar de remeter ao bolo, n√£o √© o bolo em si; √© apenas uma ideia do que ser√° o bolo -- uma **classe**. Quando voc√™ executa os passos descritos na receita o bolo se torna um **objeto** real.


Um **objeto** √© criado a partir de uma classe.

```python
class Carro:     # Define-se a "receita" para criar um 'carro'.
  odometro = 0   # Define que um 'carro' possui um 'od√¥metro'.

meu_carro = Carro()         # O objeto 'carro' ganha forma atrav√©s da classe.
print(meu_carro.odometro)   # Exibe o valor do atributo 'odometro' do objeto.
```

A fun√ß√£o `__init__()` √© um **construtor** da classe. Ele √© usado para inicializar as vari√°veis internas da classe.

```python
class Pessoa:
  def __init__(self, nome, idade):
    self.nome  = nome     # 'self' remete √† pr√≥pria classe. 'self.nome' √© um atributo.
    self.idade = idade    # 'self.idade' √© um atributo.

p1 = Pessoa("Jo√£o", 36)   # Cria um objeto com os atributos nome = 'Jo√£o' e
                          # idade = 36.

print(p1.nome)    # Exibe 'Jo√£o', o atributo 'nome' do objeto 'p1'.
print(p1.idade)   # Exibe 36, o atributo 'idade' do objeto 'p1'.
```

A fun√ß√£o `__str__()`, quando definida, fornece uma representa√ß√£o em *string* do objeto.

```python
# A classe original n√£o possui a fun√ß√£o __str__()
print(p1)   # Exibe o endere√ßo de mem√≥ria do objeto 'p1'.

class Pessoa:
  def __init__(self, nome, idade):
    self.nome  = nome
    self.idade = idade

  def __str__(self):
    return f"{self.nome}({self.idade})"

p1 = Pessoa("Jo√£o", 36)                      # Cria um objeto da nova classe.
print("Agora com a fun√ß√£o __str__():", p1)   # Exibe 'Jo√£o(36)'.
```

Uma fun√ß√£o interna √† uma classe √© chamada de **m√©todo**.

```python
class Pessoa:
  def __init__(self, nome, idade):
    self.nome  = nome
    self.idade = idade

  def saudacao(self):    # Cria um m√©todo chamado 'saudacao()'
    print("Oi, meu nome √©", self.nome, ".")

p1 = Pessoa("Jo√£o", 36)   # Cria um objeto da classe.
p1.saudacao()              # Exibe 'Oi, meu nome √© Jo√£o.'.
```

O par√¢metro `self` √© uma refer√™ncia √† inst√¢ncia da classe e √© usada para definir vari√°veis na pr√≥pria classe. O par√¢metro √© sempre o primeiro argumento nos m√©todos da classe e pode adotar outro nome diferente de `self`.

```python
class Pessoa:
  def __init__(this, nome, idade):    # 'this' √© um apelido para 'self'.
    this.nome  = nome
    this.idade = idade

  def saudacao(this):                 # Cria um m√©todo chamado 'saudacao()'
    print("Oi, meu nome √©", this.nome, ".")

p1 = Pessoa("Jo√£o", 36)   # Cria um objeto da classe.
p1.saudacao()              # Exibe 'Oi, meu nome √© Jo√£o.'.
```

√â poss√≠vel modificar **argumentos** (*vari√°veis*) dos objetos se estiverem vis√≠veis.

```python
p1 = Pessoa("Maria", 32)    # Cria um objeto da classe p1.
print(p1.idade)             # Exibe 32 (o valor do atributo 'idade').

p1.idade = 40               # Modifica o valor do atributo 'idade'.
print(p1.idade)             # Exibe 40 (o novo valor do atributo 'idade').
```

√â poss√≠vel remover **argumentos** de um objeto usando a palavra `del`.

```python
p1 = Pessoa("Maria", 32)    # Cria um objeto da classe.
del p1.idade                # Remove o atributo 'idade' do objeto p1.

#print(p1.idade)    # ERRO!
```

Tamb√©m √© poss√≠vel excluir **objetos** usando a palavra `del`.

```python
p1 = Pessoa("Maria", 32)    # Cria um objeto da classe.
del p1                     # Remove o objeto p1.

# print(p1)    # ERRO! p1 n√£o existe.
```

### Importante

* A **classe** √© uma abstra√ß√£o do objeto. Ela define os *comportamentos* para cada m√©todo, mas ela n√£o existe na mem√≥ria do computador at√© algum objeto ser criado da classe. Ela √© apenas a 'receita do bolo'.
* O **objeto** √© uma representa√ß√£o da classe na mem√≥ria do computador e √© composta por vari√°veis (chamadas *argumentos*) e fun√ß√µes (chamadas *m√©todos*). √â a realiza√ß√£o de uma receita definida pela classe.
* Um objeto criado a partir de uma classe √© *instanciado* desta classe.
* As classes podem receber de uma classe "m√£e" seus argumentos e m√©todos (*heran√ßa*).
* As classes podem ter m√©todos de mesmo nome com diferentes n√∫meros de par√¢metros (*polimorfismo*).
* As classes podem ter diferentes n√≠veis de visibilidade para seus argumentos e m√©todos (*encapsulamento*).

### Heran√ßa

* A **heran√ßa** √© a capacidade das classes receberem de outra classe base seus argumentos (*vari√°veis*) e m√©todos (*fun√ß√µes*).
* Com a heran√ßa torna-se poss√≠vel reutilizar c√≥digo j√° existente, partindo de uma classe j√° implementada.
* As classes herdeiras podem utilizar a base de c√≥digoda classe herdada.

```python
class Veiculo:
  velocidade = 0.0
  def andar():
    pass

  def parar():
    pass

# A classe 'autom√≥vel' √© um 've√≠culo'
class Automovel(Veiculo):
  marcha = "N"

carro = Automovel()
print(carro.velocidade)
print(carro.marcha)
```

```python
# Exemplo do site: https://www.programiz.com/python-programming/inheritance

class Polygon:
  def __init__(self, no_of_sides):
    self.n     = no_of_sides
    self.sides = [0 for _ in range(no_of_sides)]

  def inputSides(self, sides = None):
    if sides is None:
      self.sides = [float(input("Enter side %d: " % (i + 1))) for i in range(self.n)]
    else:
      self.sides = sides

  def dispSides(self):
    for i in range(self.n):
      print("Side %d is %f", i + 1, self.sides[i])

class Triangle(Polygon):
  def __init__(self):
    Polygon.__init__(self, 3)

  def findArea(self):
    a, b, c = self.sides
    # Calculate semi-perimete
    s = (a + b + c) / 2.0
    area = (s * (s - a) * (s - b) * (s - c)) ** 0.5
    print("The area of the triangle is %0.2f" % area)

t = Triangle()
t.inputSides([1, 2, 3])    # M√©todo na classe m√£e (Polygon)
t.findArea()               # M√©todo na classe filha (Triangle)
```

Os argumentos e m√©todos da classe m√£e sobrepostos na heran√ßa podem ser acessados com uma refer√™ncia √† classe m√£e.

```python
class Animal:
  def comer(self):
    print("Eu sei comer.")

class Cachorro(Animal):
  # O m√©todo comer de 'Cachorro' sobrep√µe o m√©todo comer de 'Animal'.
  def comer(self):
    Animal.comer(self)   # Invoca o m√©todo da classe 'Animal'.
    print("Eu adoro roer ossos.")

labrador = Cachorro()
labrador.comer()
```

### Encapsulamento

O **encapsulamento** permite definir n√≠veis de acesso para argumentos (*vari√°veis*) e m√©todos (*fun√ß√µes*).

* **public**: os elementos s√£o acess√≠veis dentro e fora da classe, inclusive em classes filhas.
* **protected**: elementos acess√≠veis dentro da classe que o define nas classes filhas.
* **private**: elementos acess√≠veis somente na classe que o define.
* No Python existem apenas atributos e m√©todos **private**; n√£o existe o conceito de **protected**.

* Argumentos **p√∫blicos** possuem nome iniciado por uma letra: `nome = 5`.

* M√©todos **p√∫blicos** possuem nome iniciado por uma letra: `def nome()`.

* Argumentos **privados** possuem nome iniciado por 2 tra√ßos baixos: `__nome = 5`.

* M√©todos **p√∫blicos** possuem nome iniciado por 2 tra√ßos baixos: `def __nome()`.


```python
class Base:
  # M√©todo p√∫blico (public)
  def publico(self):
    pass

  # M√©todo privado (private)
  def __privado(self):
    pass

class Filha(Base):
  def teste(self):
    Base.publico(self)
    #Base.__privado(self)   # ERRO! Inacess√≠vel na classe Filha.

obj1 = Base()
obj2 = Filha()

obj1.publico()      # Acess√≠vel.
#obj1.__privado()   # ERRO! Inacess√≠vel fora da classe Base.
obj2.publico()      # Acess√≠vel.
#obj2.__privado()   # ERRO! Inacess√≠vel fora da classe Filha.

obj2.teste()        # Acess√≠vel.
```
