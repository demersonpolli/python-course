## Sele√ß√£o de valores

**Nota:** Enquanto as express√µes para sele√ß√£o e atribui√ß√£o no NumPy/Python padr√£o sejam intuitivos, em c√≥digos de produ√ß√£o, √© recomendado utilizar os m√©todos otimizados do *Pandas* para acessar dados `DataFrame.at()`, `DataFrame.iat()`, `DataFrame.loc()` e `DataFrame.iloc()`.

### GetItem ([])

Para um `DataFrame`, ao receber um nome entre colchetes, a coluna correspondente √© selecionada.

## Dados faltantes (*missing data*)

Para o *NumPy*, `np.nan` representa um dado faltante. Ele √©, por padr√£o, exclu√≠do dos c√°lculos.

Reindexa√ß√£o permite mudar/adicionar/excluir o √≠ndice de um eixo especifico e retorna uma c√≥pia dos dados:

```python
df1 = df.reindex(index=datas[0:4], columns=list(df.columns) + ["E"])
df1.loc[datas[0] : datas[1], "E"] = 1
df1
```

`DataFrame.dropna()` ignora as linhas que possuem dados faltantes:

```python
df1.dropna(how="any")
```

`DataFrame.fillna()` preenche os dados faltantes com o valor fornecido:

```python
df1.fillna(value=5)
```

`isna()` retorna uma matriz l√≥gica indicando as posi√ß√µes faltantes:

```python
pd.isna(df1)
```

## Opera√ß√µes com dados faltantes

As opera√ß√µes, em geral, excluem os dados faltantes.

Exemplo, calculando a m√©dia para cada coluna:

```python
df.mean()
```

Calculando a m√©dia para cada linha:

```python
df.mean(axis=1)
```

As opera√ß√µes que envolvam outras `Series` ou `DataFrame` com √≠ndices ou colunas diferentes ir√£o alinhar os resultados com a uni√£o dos √≠ndices e nomes de colunas. Al√©m disso, o *Pandas* automaticamente propaga os valores ao longo das dimens√µes especificadas e preenche os pares n√£o alinhados com `np.nan`.

## Agrupamento

Por ‚Äúagrupamento‚Äù se refere a um processo que envolve um ou mais dos seguintes passos:

- **particionar** os dados em grupos basedos em algum crit√©rio;

- **aplicar** uma fun√ß√£o em cada grupo independentemente;

- **combinar** os resultados em uma estrutura de dados.

```python
df = pd.DataFrame(
    {
        "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo"],
        "B": ["um", "um", "dois", "tr√™s", "dois", "dois", "um", "tr√™s"],
        "C": np.random.randn(8),
        "D": np.random.randn(8),
    }
)

df
```

Exemplo, agrupando por uma coluna, selecionando outras colunas e aplicando a fun√ß√£o `sum()` aos dados resultantes:

```python
df.groupby("A")[["C", "D"]].sum()
```

Agrupar por m√∫ltiplas colunas forma um `MultiIndex`.

```python
df.groupby(["A", "B"]).sum()
```

## Reorganiza√ß√£o

### Empilhamento

```python
arrays = [
   ["bar", "bar", "baz", "baz", "foo", "foo", "qux", "qux"],
   ["um", "dois", "um", "dois", "um", "dois", "um", "dois"],
]


index = pd.MultiIndex.from_arrays(arrays, names=["first", "second"])

df = pd.DataFrame(np.random.randn(8, 2), index=index, columns=["A", "B"])
df2 = df[:4]
df2
```

O m√©todo `stack()` "comprime" um n√≠vel das colunas do `DataFrame`:

```python
stacked = df2.stack()
stacked
```

A opera√ß√£o contr√°ria do empilhamento √© realizado por `unstack()` que, por padr√£o, desempilha pelo √∫ltimo n√≠vel do `MultiIndex`:

```python
stacked.unstack()
```

```python
stacked.unstack(1)
```

```python
stacked.unstack(0)
```

### Pivot tables

Uma *pivot table* √© uma tabela de resumo de uma tabela original.

```python
df = pd.DataFrame(
    {
        "A": ["um", "um", "dois", "tr√™s"] * 3,
        "B": ["A", "B", "C"] * 4,
        "C": ["foo", "foo", "foo", "bar", "bar", "bar"] * 2,
        "D": np.random.randn(12),
        "E": np.random.randn(12),
    }
)

df
```

`pivot_table()` pivoteia um `DataFrame` especificando *valores* (`values`), *√≠ndices* (`index`) e *colunas* (`columns`).

```python
pd.pivot_table(df, values="D", index=["A", "B"], columns=["C"])
```

## S√©ries temporais

*Pandas* possue funcionalidades simples, poderosas e eficientes para fazer opera√ß√µes de reamostragem para convers√£o de frequ√™ncias (ex: converter segundos para um escala de 5 minutos). Isto √© bastante comum em, mas n√£o limitado a, aplica√ß√µes financeiras.

```python
ts_utc = ts.tz_localize("UTC")
ts_utc
```

```python
ts_utc.tz_convert("America/Sao_Paulo")
```

## Dados categorizados

*Pandas* permite a defini√ß√£o de dados categorizados. √â o equivalente ao `factor` na linguagem [R](https://www.r-project.org).

```python
df = pd.DataFrame(
    {"id": [1, 2, 3, 4, 5, 6], "nota": ["a", "b", "b", "a", "a", "e"]}
)
```

Convertendo as notas em texto para categorizado:

```python
df["grau"] = df["nota"].astype("category")

df["grau"]
```

Renomeando as categorias para nomes com mais significado:

```python
new_categories = ["muito bom", "bom", "muito ruim"]

df["grau"] = df["grau"].cat.rename_categories(new_categories)
```

Reorganizando as categorias e simultaneamente adicionando as categorias faltantes (m√©todos em `Series.cat()` retornam uma nova `Series` por padr√£o):

```python

df["grau"] = df["grau"].cat.set_categories(
    ["muito ruim", "ruim", "regular", "bom", "muito bom"]
)

df["grau"]
```

A  ordena√ß√£o nas categorias √© por classe, n√£o alfab√©tica:

```python
df.sort_values(by="grau")
```

Agrupamento de vari√°veis categorizadas com `observed=False` tamb√©m mostram as categorias vazias:

```python
df.groupby("grau", observed=False).size()
```

## Importando e exportando dados

### Valores separados por v√≠rgula (CSV)

Para escrever um arquivo CSV (*valores separados por v√≠rgula*) use `DataFrame.to_csv()`:

```python
df = pd.DataFrame(np.random.randint(0, 5, (10, 5)))
df.to_csv("foo.csv")
```

Para ler o arquivo CSV use `DataFrame.read_csv()`:

```python
pd.read_csv("foo.csv")
```

### Parquet

**Parquet** √© um formato bastante eficiente para armazenar grandes volumes de dados que ser√£o consultados por colunas. Para escrever um arquivo *Parquet* use `DataFrame.to_parquet()`:

```python
# O Parquet exige que as colunas tenham nomes em string.
df2 = df.set_axis(['A','B','C','D','E'], axis= 1)
df2.to_parquet("foo.parquet")
```

Para ler um arquivo *Parquet* use `DataFrame.read_parquet()`:

```python
pd.read_parquet("foo.parquet")
```

### Excel

Para escrever um arquivo Excel use `DataFrame.to_excel()`:

```python
df.to_excel("foo.xlsx", sheet_name="Sheet1")
```

Para ler um arquivo Excel use `DataFrame.read_excel()`:

```python
pd.read_excel("foo.xlsx", "Sheet1", index_col=None, na_values=["NA"])
```
